# Descriptive Statistics

```python
# Jaguar vs Panther
resistance = np.array([
    [10.6, 9.1, 9.3, 9.8, 10.5, 10.4, 9.5, 11, 10.4, 3, 9.8],
    [10.1, 11, 9.1, 20, 9.2, 10.8, 9.9, 9.2, 9.1, 9.1, np.nan]
    ]).T
np.shape(resistance)
```

```python
df = pd.DataFrame(resistance, columns=['Jaguar', 'Panther'])
df 
```

```python
type(df)
```

```python
df['Panther']
```

```python
df.fillna(100)
```

```python
df
```

```python
df.describe()
```

```python
# Median
jag_median = df['Jaguar'].median()
pan_median = df['Panther'].median()

print(f'Jaguar Median = {jag_median:.2f}')
print(f'Panther Median = {pan_median:.2f}')
```

```python
# Mode
jag_mode = df['Jaguar'].mode()
pan_mode = df['Panther'].mode()
```

```python
jag_mode
```

```python
type(jag_mode)
```

```python
print('Jaguar Mode')

for i in jag_mode:
    print(i)
```

```python
print('Panther Mode')

for i in pan_mode:
    print(i)
```

```python
# Jaguar IQR

Q1 = df['Jaguar'].quantile(0.25)
Q3 = df['Jaguar'].quantile(0.75)

print(f'Jaguar Q1 = {Q1:.2f}')
print(f'Jaguar Q3 = {Q3:.2f}')
print(f'Jaguar IQR = {Q3 - Q1:.2f}')
```

```python
# Panth IQR

Q1 = df['Panther'].quantile(0.25)
Q3 = df['Panther'].quantile(0.75)

print(f'Panther Q1 = {Q1:.2f}')
print(f'Panther Q3 = {Q3:.2f}')
print(f'Panther IQR = {Q3 - Q1:.2f}')
```

```python
jag_mean = df['Jaguar'].mean()
pan_mean = df['Panther'].mean()

plt.Figure()
bars = plt.bar(['Jaguar', 'Panther'], [jag_mean, pan_mean], color=['blue', 'darkblue'])

plt.title('Mean Resistance of Jaguar vs Panther')
plt.ylabel('Resistance (ohms)')

plt.bar_label(bars, padding=5)
plt.margins(y=0.5)
plt.show()
```
