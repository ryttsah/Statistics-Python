```python
import numpy as np
from scipy import stats
```

# Measures of Central Tendencies

```python
# Jaguar vs Panther
resistance = np.array([
    [10.6, 9.1, 9.3, 9.8, 10.5, 10.4, 9.5, 11, 10.4, 3, 9.8],
    [10.1, 11, 9.1, 20, 9.2, 10.8, 9.9, 9.2, 9.1, 9.1, np.nan]
    ])
np.shape(resistance)
```

```python
res_mean = np.nanmean(resistance, axis=1)
print("Jaguar Mean: " + str((res_mean[0]).round(2)))
print("Panther Mean: " + str((res_mean[1]).round(2)))
```

```python
res_media = np.nanmedian(resistance, axis=1)
print("Jaguar Median: " + str((res_mean[0]).round(2)))
print("Panther Median: " + str((res_mean[1]).round(2)))
```

```python
res_mode = stats.mode(resistance, axis=1)
res_mode
```

```python
from statistics import multimode


jag_res = resistance[0]

print("Jaguar Mode: ")
for i in multimode(jag_res):
    print(str(i) + " Ohms")
```

```python
panth_res = resistance[0]

print("Jaguar Mode: ")
for i in multimode(panth_res):
    print(str(i) + " Ohms")
```

```python
res_mode = stats.mode(resistance, axis=1)
print("Jaguar Mode: " + str((res_mean[0]).round(2)))
print("Panther Mode: " + str((res_mean[1]).round(2)))
```

# Measures of Range, Variance, SD

```python
# Range
res_max = np.nanmax(resistance, axis = 1)
res_min = np.nanmin(resistance, axis = 1)

res_range = res_max - res_min
res_range
```

```python
print("Jaguar Mode: " + str((res_range[0]).round(2)))
print("Panther Mode: " + str((res_range[1]).round(2)))
```

```python
res_q1 = np.nanpercentile(resistance, 25, axis = 1)
print("Jaguar Q1: " + str((res_q1[0]).round(2)))
print("Panther Q1: " + str((res_q1[1]).round(2)))
```

```python
res_q3 = np.nanpercentile(resistance, 75, axis = 1)
print("Jaguar Q1: " + str((res_q3[0]).round(2)))
print("Panther Q1: " + str((res_q3[1]).round(2)))
```

```python
print("Jaguar IQR: " + str((res_q3[0]-res_q1[0]).round(2)))
print("Panther IQR: " + str((res_q3[1]-res_q1[1]).round(2)))
```

```python
res_var = np.nanvar(resistance, ddof = 1, axis = 1)
print("Jaguar Variance: " + str((res_var[0]).round(2)))
print("Panther Variance: " + str((res_var[1]).round(2)))
```

```python
res_std = np.nanstd(resistance, ddof = 1, axis = 1)
print("Jaguar Standard Deviation: " + str((res_std[0]).round(2)))
print("Panther Standard Deviation: " + str((res_std[1]).round(2)))
```
